```python
# 2. Define the columns that look like numbers but are actually categories
cat_cols = [
    'Education', 
    'EnvironmentSatisfaction', 
    'JobInvolvement', 
    'JobLevel', 
    'JobSatisfaction', 
    'PerformanceRating', 
    'RelationshipSatisfaction', 
    'StockOptionLevel', 
    'WorkLifeBalance'
]

# 3. Convert them to 'object' type
# This tells Python: "Treat these as labels/strings, not math numbers"
data[cat_cols] = data[cat_cols].astype('object')

# 5. Verify the changes
print(data.info())
```

```python
Q1 = data['MonthlyIncome'].quantile(0.25)
Q3 = data['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1
print("Q1 values:", Q1)
print("Q3 values:", Q3)
print("IQR values:", IQR)
```

```python
lower_limit = Q1 - 1.5 * (IQR)
upper_limit = Q3 + 1.5 * (IQR)
print(lower_limit)
print(upper_limit)
```

```python
outliers = np.where((data['MonthlyIncome'] > upper_limit) | (data['MonthlyIncome'] < lower_limit))
outliers
outliers = ((data['MonthlyIncome'] > upper_limit) | (data['MonthlyIncome'] < lower_limit)).sum()
outliers
```

```python
import seaborn as sns
sns.boxplot(data.MonthlyIncome)
```
